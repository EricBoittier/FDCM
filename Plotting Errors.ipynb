```python
import os, sys
import matplotlib.pyplot as plt
plt.style.use(['science','ieee', "no-latex", "std-colors"])

def plot_transformation_errors(name):
    f = open(name).readlines()
    totals = [float(x.split()[1]) for x in f if x.__contains__("Total        ")]
    phi = list(range(-180, 160, 20))
    plt.plot(phi,totals, label="Transformation")
    plt.grid()
    plt.ylabel("ESP Error (kcal mol$^{-1}$ $e^{-1}$)")
    plt.xlabel("$\Phi~(^{\circ})  $")
    plt.ylim(0, 5)
    
def plot_optimised_errors(name):
    f = open(name).readlines()
    totals = [float(x.split()[1]) for x in f if x.__contains__("Error       ")]
    phi = list(range(-180, 160, 20))
    plt.plot(phi,totals, label="Transformation + 100 steps GD" )
#     plt.grid()
    plt.ylabel("ESP Error (kcal mol$^{-1}$ $e^{-1}$)")
    plt.xlabel("$\Phi~(^{\circ})  $")
    plt.ylim(0, 5)
```

```python
n = "ester"
name = f"{n}_transformed.errors"
plot_transformation_errors(name)
#plot_optimised_errors(name)
name = f"{n}_100.errors"
# plot_transformation_errors(name)
plot_optimised_errors(name)
plt.legend()
plt.savefig(f"{n}_MDCM_errors.pdf", bbox_inches="tight")
```
